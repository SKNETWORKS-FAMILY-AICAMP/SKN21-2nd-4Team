```python
import matplotlib.pyplot as plt
import numpy as np

# 모델 이름
models = ["DecisionTree", "RandomForest", "linearRegression", "LGBM", "XGBoost"]

# 성능 지표
accuracy =  [0, 0.84, 0, 0.85, 0.8540601930721181]
f1 =        [0, 0.75, 0, 0.75, 0.7512100677637947]
recall =    [0, 0.87, 0, 0.83, 0.8308351177730193]
precision = [0, 0.66, 0, 0.68, 0.6855123674911661]

x = np.arange(len(models))
width = 0.2

plt.figure(figsize=(12, 6))

plt.bar(x - width*1.5, accuracy, width, label='Accuracy')
plt.bar(x - width*0.5, f1, width, label='F1-score')
plt.bar(x + width*0.5, recall, width, label='Recall')
plt.bar(x + width*1.5, precision, width, label='Precision')

plt.xticks(x, models)
plt.ylabel("Score")
plt.title("Model Performance Comparison (Best Version per Model)")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```
