```python
import pandas as pd
import os

paths = [
    "/Users/anhyebin/Documents/SKN21/SKN21-2nd-4Team/data/raw/Customer_Churn_Dataset_0_impute.csv",
    "/Users/anhyebin/Documents/SKN21/SKN21-2nd-4Team/data/raw/Customer_Churn_Dataset_knn.csv",
    "/Users/anhyebin/Documents/SKN21/SKN21-2nd-4Team/data/raw/Customer_Churn_Dataset_simple_frequent.csv",
    "/Users/anhyebin/Documents/SKN21/SKN21-2nd-4Team/data/raw/Customer_Churn_Dataset_simple_median.csv"
]

output_path = '/Users/anhyebin/Documents/SKN21/SKN21-2nd-4Team/data/processed'
```

```python
#범주형 컬럼 매핑
binary_mapping = {
    'Yes': 1,
    'No': 0,
    'Male': 1,
    'Female': 0,
    'No internet service': 0,
    'No phone service': 0
}


# #범주형 컬럼
categorical_cols = [
    'gender', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 
    'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract', 
    'PaperlessBilling', 'PaymentMethod', 'Churn'
]


for path in paths:
    df = pd.read_csv(path)


    df = df.drop(columns = ["customerID"])

    for col in categorical_cols:
        df[col] = df[col].map(binary_mapping).fillna(df[col])

    

    file_name = os.path.basename(path).replace(".csv", "_baseline.csv")
    save_path = os.path.join(output_path, file_name)


    df.to_csv(save_path,index = False)
    print(f"{save_path} 저장 완료")
```
