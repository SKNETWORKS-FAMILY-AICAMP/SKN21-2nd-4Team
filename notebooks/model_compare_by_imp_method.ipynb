```python
import matplotlib.pyplot as plt
import numpy as np
```

1. KNN

```python

# 모델 이름
models = ["DecisionTree", "RandomForest", "LogisticRegression", "XGBoost", "SVM"]

# 성능 지표
accuracy = [0.8364565587734242, 0.8512208972174901, 0.858603066439523, 0.8540601930721181, 0.8642816581487791]
f1 = [0.6749435665914221, 0.6981566820276498, 0.7205387205387206, 0.7512100677637947, 0.7311586051743532]
recall = [0.6402569593147751, 0.6488222698072805, 0.6873661670235546, 0.8308351177730193, 0.69593147751606]
precision = [0.7136038186157518, 0.7556109725685786, 0.7570754716981132, 0.6855123674911661, 0.7701421800947867]
print(f"f1 max : {models[f1.index(max(f1))], max(f1)}")
print(f"recall max: {models[recall.index(max(recall))], max(recall)}")
x = np.arange(len(models))
width = 0.2
plt.figure(figsize=(12, 6))

plt.bar(x - width*1.5, accuracy, width, label='Accuracy')
plt.bar(x - width*0.5, f1, width, label='F1-score')
plt.bar(x + width*0.5, recall, width, label='Recall')
plt.bar(x + width*1.5, precision, width, label='Precision')

plt.xticks(x, models)
plt.ylabel("Score")
plt.title("Model Performance Comparison (Best Version per Model)")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```
